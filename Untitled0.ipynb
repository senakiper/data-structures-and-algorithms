# Implement the function myCurve() that works similar to the R function curve().
# The input should be an expression that depends on x, and a lower and upper bound that defines the interval on which x should be varied.

myCurve <- function(expr, lower, upper) {
  # Create a sequence of values for x from lower to upper
  x_vals <- seq(lower, upper, length.out = 100)
  
  # Evaluate the expression for each value of x
  y_vals <- sapply(x_vals, function(x) eval(expr))
  
  # Plot the results
  plot(x_vals, y_vals, type = "l", xlab = "x", ylab = "f(x)", main = "Plot of the Function")
}

# a) Write a function that receives an arbitrary expression as input. Find out if the constants T resp. F are being used. If yes, replace them by TRUE and FALSE. Return the modified expression.

replace_TF <- function(expr) {
  # Check if T or F are used and replace them
  expr <- gsub("\\bT\\b", "TRUE", expr)
  expr <- gsub("\\bF\\b", "FALSE", expr)
  return(expr)
}

# b) Perform the replacement only if T or F is not on the left side of the assignment.

replace_TF_conditional <- function(expr) {
  # Split the expression into left-hand side and right-hand side
  lhs <- deparse(substitute(expr[[2]]))  # Get the left-hand side of the assignment
  rhs <- deparse(substitute(expr[[3]]))  # Get the right-hand side of the assignment

  # Replace T and F on the right-hand side only
  if (!grepl("^[a-zA-Z]", lhs)) {
    rhs <- gsub("\\bT\\b", "TRUE", rhs)
    rhs <- gsub("\\bF\\b", "FALSE", rhs)
  }
  # Reconstruct the expression
  expr[[3]] <- rhs
  return(expr)
}

# a) How many standard normally distributed random variables does one have to draw until the sum of the squared random variables is bigger than a constant?

# Using a for loop
draw_for <- function(const) {
  sum_sq <- 0
  count <- 0
  for (i in 1:10000) {
    count <- count + 1
    sum_sq <- sum_sq + rnorm(1)^2
    if (sum_sq > const) {
      return(count)
    }
  }
}

# Using a while loop
draw_while <- function(const) {
  sum_sq <- 0
  count <- 0
  while (sum_sq <= const) {
    count <- count + 1
    sum_sq <- sum_sq + rnorm(1)^2
  }
  return(count)
}

# Using a repeat loop
draw_repeat <- function(const) {
  sum_sq <- 0
  count <- 0
  repeat {
    count <- count + 1
    sum_sq <- sum_sq + rnorm(1)^2
    if (sum_sq > const) {
      break
    }
  }
  return(count)
}

sum_for <- function(x) {
  total <- 0
  for (i in 1:length(x)) {
    total <- total + x[i]
  }
  return(total)
}

sum_while <- function(x) {
  total <- 0
  i <- 1
  while (i <= length(x)) {
    total <- total + x[i]
    i <- i + 1
  }
  return(total)
}

# a) Write a function that, depending on the supplied character type, returns either the arithmetic mean, the median, the standard deviation, or the MAD (median absolute difference) of a vector x.

calculate_stat <- function(x, stat_type) {
  result <- switch(stat_type,
                   "mean" = mean(x),
                   "median" = median(x),
                   "sd" = sd(x),
                   "mad" = mad(x),
                   stop("Invalid stat_type"))
  return(result)
}

# b) For coding/decoding of a text with the Caesar cipher.

caesar_cipher <- function(text, shift, decode = FALSE) {
  # Convert text to lowercase and split into individual characters
  text <- tolower(text)
  alphabet <- letters
  
  # If decoding, adjust the shift to be negative
  if (decode) {
    shift <- -shift
  }
  
  # Apply Caesar cipher shift
  ciphered_text <- sapply(strsplit(text, NULL)[[1]], function(char) {
    if (char %in% alphabet) {
      new_pos <- (which(alphabet == char) + shift - 1) %% 26 + 1
      return(alphabet[new_pos])
    } else {
      return(char)  # Keep non-alphabet characters unchanged
    }
  })
  
  return(paste(ciphered_text, collapse = ""))
}
